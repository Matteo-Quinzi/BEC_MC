```python
import numpy as np
import matplotlib.pyplot as plt

data = np.genfromtxt('DMC.out', skip_header=1)
r = data[:,0]
rho = data[:,1]
eigenvec = data[:,2]


fig, ax = plt.subplots()
#ax.set_xlim(0.,1.)
#ax.set_ylim(0.,10)
ax.plot(r,rho/max(rho)/r)
ax.plot(r,eigenvec/max(eigenvec))
```
