```python
import numpy as np
import matplotlib.pyplot as plt

data = np.genfromtxt('DMC.out', skip_header=1)
r = data[:,0]
rho = data[:,1]

data = np.genfromtxt('eigenvec.out')
eigenvec = data[:,1]

fig, ax = plt.subplots()
ax.plot(r,rho/r**2.0)
ax.plot(r,eigenvec)
```
